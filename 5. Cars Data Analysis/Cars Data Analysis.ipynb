```python
# first we'll have to rename the 'MSRP' column to 'MSRP ($)' and 'Invoice' to 'Invoice ($)'.
car.rename(columns = {'MSRP' : 'MSRP ($)'}, inplace = True)
```

```python
car.rename(columns = {'Invoice' : 'Invoice ($)'}, inplace = True)
```

```python
car.head()
```

```python
car.info()
```

```python
# now we have to remove the dollar sign and spaces and change the datatype of the 'MSRP ($)' and 'Invoice ($)' columns
car["MSRP ($)"] = car["MSRP ($)"].replace(r'[\$,]', '', regex=True).astype(float)
car["Invoice ($)"] = car["Invoice ($)"].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
car.head()
```

```python
car.info()
```

```python
# now to fill all null values with the mean of their respective column:
car = car.fillna(car.mean(numeric_only = True)) #only applicable to the int and float datatype columns
```

```python
# histograms for the dataframe:
import matplotlib.pyplot as plt
```

```python
# list of columns to plot
numeric_column = ['MSRP ($)', 'EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length']

# plot histograms for each numn=eric column
for col in numeric_column:
    plt.figure(figsize = (8, 4))
    car[col].dropna().hist(bins = 20, edgecolor = 'black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

```python
# Grouped Histogram of Horsepower by type:
import seaborn as sns

plt.figure(figsize = (10, 6))
sns.histplot(data = car, x = 'Horsepower', hue = 'Type', element = 'step', stat = 'count', common_norm = False, bins = 20)
plt.title('Horsepower distribution by Car type')
plt.xlabel('Horsepower')
plt.ylabel('Coount')
plt.show()
```

```python
# Horsepower distribution by origin

plt.figure(figsize = (10, 6))
sns.histplot(data = car, x = 'Horsepower', hue = 'Origin', element = 'step', stat = 'density', common_norm = False, bins = 30)
plt.title('Horsepower distribution by Origin')
plt.xlabel('Horsepower')
plt.ylabel('Density')
plt.show()
```

```python
# Boxplor: MPG_City by Drivetrain

plt.figure(figsize = (8, 6))
sns.boxplot(data = car, x = 'DriveTrain', y = 'MPG_City')
plt.title('City MPG by DriveTrain')
plt.xlabel('DriveTrain')
plt.ylabel('MPG City')
plt.show()
```
